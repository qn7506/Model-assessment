```python
poly = PolynomialFeatures(degree=2)
displacement = poly.fit_transform(data.iloc[:,2:3])
displacement = pd.DataFrame(displacement[:,1:], columns = ['displacement1','displacement2'])
horsepower = poly.fit_transform(data.iloc[:,3:4])
horsepower = pd.DataFrame(horsepower[:,1:], columns = ["horsepower1", "horsepower2"])
weight = poly.fit_transform(data.iloc[:,4:5])
weight = pd.DataFrame(weight[:,1:], columns = ["weight1", "weight2"])

log_displacement = np.log(data.iloc[:,2:3])
log_displacement = log_displacement.rename(columns={"displacement":"log_displacement"})
log_horsepower = np.log(data.iloc[:,3:4])
log_horsepower = log_horsepower.rename(columns={"horsepower":"log_horsepower"})
log_weight = np.log(data.iloc[:,4:5])
log_weight = log_weight.rename(columns={"weight":"log_weight"})
log_mpg = np.log(data.iloc[:,0:1])
log_mpg = log_mpg.rename(columns={"mpg":"log_mpg"})

Z, fit_lambda = stats.boxcox(data5.mpg)
Z = pd.DataFrame(Z, columns=['Z'])

data_new = pd.concat([Z,log_mpg,data.iloc[:,0:2],displacement,horsepower,weight,log_displacement,log_horsepower,log_weight,data.iloc[:,5]], axis = 1)
data_new
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
error1 = []
error2 = []
error3 = []
error4 = []
error5 = []
for i in range(1,101):
    x = data_new.iloc[:,3:]
    y = data_new.iloc[:,:3]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)

    ols = sm.OLS(y_train[["mpg"]],sm.add_constant(x_train[["cylinders","displacement1","displacement2",
                                                           "horsepower1","horsepower2","weight1","weight2","acceleration"]]))
    lm1 = ols.fit()
    y_pred1 = lm1.predict(sm.add_constant(x_test[["cylinders","displacement1","displacement2","horsepower1",
                                                  "horsepower2","weight1","weight2","acceleration"]]))
    error1.append(mean_squared_error(y_pred1,y_test[["mpg"]]))

    ols = sm.OLS(y_train[["mpg"]], sm.add_constant(x_train[["cylinders","log_displacement","log_horsepower","log_weight","acceleration"]]))
    lm2 = ols.fit()
    y_pred2 = lm2.predict(sm.add_constant(x_test[["cylinders","log_displacement","log_horsepower","log_weight","acceleration"]]))
    error2.append(mean_squared_error(y_pred2,y_test[["mpg"]]))

    ols = sm.OLS(y_train[["log_mpg"]], sm.add_constant(x_train[["cylinders","displacement1","horsepower1","weight1","acceleration"]]))
    lm3 = ols.fit()
    y_pred3 = np.exp(lm3.predict(sm.add_constant(x_test[["cylinders","displacement1","horsepower1","weight1","acceleration"]])))
    error3.append(mean_squared_error(y_pred3,y_test[["mpg"]]))

    ols = sm.OLS(y_train[["log_mpg"]], sm.add_constant(x_train[["cylinders","log_displacement","log_horsepower","log_weight","acceleration"]]))
    lm4 = ols.fit()
    y_pred4 = np.exp(lm4.predict(sm.add_constant(x_test[["cylinders","log_displacement","log_horsepower","log_weight","acceleration"]])))
    error4.append(mean_squared_error(y_pred4,y_test[["mpg"]]))

    ols = sm.OLS(y_train[["Z"]], sm.add_constant(x_train[["cylinders","log_displacement","log_horsepower","log_weight","acceleration"]]))
    lm5 = ols.fit()
    y_pred5 = (lm5.predict(sm.add_constant(x_test[["cylinders","log_displacement","log_horsepower",
                                                   "log_weight","acceleration"]]))*fit_lambda+1)**(1/fit_lambda)
    error5.append(mean_squared_error(y_pred5,y_test[["mpg"]]))
```
